Task:
- Predict the percapita income of canada in 2020 using linear regression.

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
# Rename the column for easier access
df = df.rename(columns={'per capita income (US$)': 'income'})

# Then plot
plt.xlabel('years')
plt.ylabel('income')
plt.scatter(df['year'], df['income'], color='red', marker='+')
```

```python
new_df = df.drop('income',axis='columns')
new_df
```

```python
income = df.income
income
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,income)
```

```python
reg.predict([[2020]])
```
